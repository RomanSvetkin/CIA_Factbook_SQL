```python
%%sql
SELECT *
  FROM facts
 LIMIT 5;
```

```python
%%sql
SELECT MIN(population) AS min_pop,
       MAX(population) AS max_pop,
       MIN(population_growth) AS min_pop_growth,
       MAX(population_growth) AS max_pop_growth
  FROM facts
```

```python
%%sql
SELECT name, population      
  FROM facts
 ORDER BY population DESC
 LIMIT 5;
```

```python
%%sql
SELECT name, population      
  FROM facts
 ORDER BY population
 LIMIT 50;
```

```python
%%sql
SELECT MIN(population) AS min_pop,
       MAX(population) AS max_pop,
       MIN(population_growth) AS min_pop_growth,
       MAX(population_growth) AS max_pop_growth
  FROM facts
 WHERE name <> 'World';
  
```

```python
%%sql
SELECT ROUND(AVG(population), 2) AS avg_pop, ROUND(AVG(area), 2) AS avg_area
  FROM facts
 WHERE name <> 'World';
```

```python
%%sql
SELECT name, population, area
  FROM facts
 WHERE population > (SELECT AVG(population)
                     FROM facts) AND 
                     area < (SELECT AVG(area)
                     FROM facts
                    );
```

```python
%%sql
SELECT *
  FROM facts
 WHERE population > (SELECT AVG(population)
                    FROM facts)
 ORDER BY population DESC;
```

```python
%%sql
SELECT *
  FROM facts
 WHERE population > (SELECT AVG(population_growth)
                    FROM facts)
 ORDER BY population_growth DESC;
```

```python
%%sql
SELECT name, area_land, area_water, ROUND(CAST(area_water AS FLOAT) / area_land, 2) AS water_to_land
  FROM facts
 ORDER BY water_to_land DESC;
```

```python
%%sql
SELECT name, area_land, area_water, population, population_growth, ROUND(CAST(population AS FLOAT) * (population_growth) - population, 2) AS most_pop_added
  FROM facts
 ORDER BY most_pop_added DESC;
```

```python
%%sql
SELECT name, population, birth_rate, death_rate, ROUND(CAST(death_rate AS FLOAT) / birth_rate, 2) AS death_birth_ratio
  FROM facts
  WHERE death_rate > birth_rate
 ORDER BY death_birth_ratio DESC;
```

```python
%%sql
SELECT name, population, area, ROUND(CAST(population AS FLOAT) / area, 2) AS pop_area_ratio
  FROM facts
  WHERE death_rate > birth_rate
 ORDER BY pop_area_ratio DESC;
```
